```python
import numpy as np #; np.random.seed(0)
import matplotlib.pyplot as plt
```

```python
state_history = []
state = 100
states = np.arange(201)
final_states = [states[0], states[-1]]
p = 0.51
q = 1 - p
N_STATES = final_states[1] - final_states[0] + 1
```

```python
P = np.zeros((N_STATES, N_STATES))
for i in range(N_STATES):
    if i not in final_states: 
        P[i,i+1] = p
        P[i,i-1] = q
    elif i == final_states[0]: 
        P[i,i+1] = p
        P[i,i] = q
    elif i == final_states[1]: 
        P[i,i] = p
        P[i,i-1] = q
#print(P)
```

```python
while True:
    next_state = np.random.choice(states, p=P[state,:])
    prob = P[state, next_state]
    #print(f"현재 스테이트: {state}, 넥스트 스테이트: {next_state}, 확률: {prob}")
    #print("현재 스테이트: {}, 넥스트 스테이트: {}, 확률: {}".format(state,next_state,prob))
    state = next_state
    state_history.append(state)
    if state == final_states[0]:
        #print("돈을 다 잃었습니다. ㅠㅠ")
        break
    elif state == final_states[1]:
        #print("목표달성. :)")
        break 
```

```python
plt.plot(state_history)
plt.show()
```
