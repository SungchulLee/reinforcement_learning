```python
from dqn_policy import POLICY
from model_MDP import MDP


print("Print Policy")
policy = POLICY(policy_choice='optimal')
print(policy.policy)
print()


print("Run one simulation")
env = MDP(policy.policy)
env.current_state, done = env.reset()
while not done:
    action, prob = env.action()
    reward, next_state, done, info, final_reward = env.step(action, prob)
    msg = "s: {:2}, a: {}, r: {:5.2f}, s1: {:2}, done: {:1}, info: {}, final_reward: {}" 
    print(msg.format(env.current_state, action, reward, next_state, \
                     done, info, final_reward))
    
    env.current_state = next_state 
```

```python
import numpy as np
import matplotlib.pyplot as plt

from dqn_policy import POLICY
from model_MDP import MDP


num_simulations = 3000


print("Print Policy")
policy = POLICY(policy_choice='bad')
print(policy.policy)
print()


print("Run many simulation")
env = MDP(policy.policy)
simulation_history = []
for _ in range(num_simulations):
    env.current_state, done = env.reset()
    while not done:
        action, prob = env.action()
        reward, next_state, done, info, final_reward = env.step(action, prob)
        env.current_state = next_state
        
#     if final_reward == 1.0:
#         simulation_history.append(1.0)
#     if final_reward == -1.0:
#         simulation_history.append(0.0)
    simulation_history.append((final_reward+1) / 2.)
    
# compute running success rate
x = np.array(simulation_history)
running_success_rate = np.cumsum(x) / (np.arange(num_simulations) + 1)

# print overall success rate
print("Number of simulations : {}".format(num_simulations))
print("Success rate          : {}".format(running_success_rate[-1]))
        
# plot running success rate
plt.plot(running_success_rate)
plt.title("Running Success Rate")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import gym

num_simulations = 3000

env = gym.make("FrozenLake-v0")

simulation_history = []
for _ in range(num_simulations):
    env.reset()
    done = False
    while not done:
        action = env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        if done:
            # win: reward = 1.0
            # lose: reward = 0.0
            simulation_history.append(reward)
            
# compute running success rate
x = np.array(simulation_history)
running_success_rate = np.cumsum(x) / (np.arange(num_simulations) + 1)

# print overall success rate
print("Number of simulations : {}".format(num_simulations))
print("Success rate          : {}".format(running_success_rate[-1]))
        
# plot running success rate
plt.plot(running_success_rate)
plt.title("Running Success Rate")
plt.show()
```
