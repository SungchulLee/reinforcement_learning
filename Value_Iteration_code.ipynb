# 1 Value Iteration for V_optimal

For action-independent reward $R$ (this is implemented)
$$V^{*}(s)=\max_{a}\left(R(s)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')\right)$$

For action-dependent reward $R$
$$V^{*}(s)=\max_{a}\left(R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')\right)$$

For action-independent reward $R$ (this is implemented)
$$Q^{*}(s,a)=R(s)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')$$

For action-dependent reward $R$
$$Q^{*}(s,a)=R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')$$

$$
\pi^{*}(s)=\mbox{argmax}_{a}Q^{*}(s,a)
$$

```python
from model_MDP import VI_V


" Value Iteration "
vi_v = VI_V()
V, _, Q, optimal_policy = vi_v.run_value_iteration()
print(V)
print()
print(Q)
print()
print(optimal_policy)


" Plot of Values "
vi_v.plot()
```

# 2 Value Iteration for Q_optimal

For action-independent reward $R$ (this is implemented)
\begin{array}{lll}
Q^{*}(s,a)&=&R(s)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')\\
&=&R(s)+\gamma\sum_{s'}P(s,a,s')\max_{a'}Q^{*}(s',a')
\end{array}

For action-dependent reward $R$
\begin{array}{lll}
Q^{*}(s,a)&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{*}(s')\\
&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')\max_{a'}Q^{*}(s',a')
\end{array}

$$
\pi^{*}(s)=\mbox{argmax}_{a}Q^{*}(s,a)
$$

```python
from model_MDP import VI_Q


" Value Iteration "
vi_q = VI_Q()
Q, _, optimal_policy = vi_q.run_value_iteration()
print(Q)
print()
print(optimal_policy)


" Plot of Values "
vi_q.plot()
```
