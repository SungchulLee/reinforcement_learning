# 1 Iterative Policy Evaluation for V

For deterministic policy $\pi$ and action-independent reward $R$
$$V^{\pi}(s)=R(s)+\gamma\sum_{s'}P(s,\pi(s),s')V^{\pi}(s')$$

For non-deterministic policy $\pi$ and action-independent reward $R$ (this is implemented)
$$V^{\pi}(s)=\sum_{a}\pi(s)\left[R(s)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\right]$$

For deterministic policy $\pi$ and action-dependent reward $R$
$$V^{\pi}(s)=R(s,\pi(s))+\gamma\sum_{s'}P(s,\pi(s),s')V^{\pi}(s')$$

For non-deterministic policy and $\pi$ action-dependent reward $R$
$$V^{\pi}(s)=\sum_{a}\pi(s,a)\left[R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\right]$$

```python
from dqn_policy import POLICY
from model_MDP import IPE_V


" Policy "
policy = POLICY(policy_choice='random')
print(policy.policy)
print()


" Iterative_Policy_Evaluation for V "
ipe = IPE_V(policy.policy, num_iteration=500)
ipe.run_ipe()
print(ipe.V)
print()


" Plot of V_history "
ipe.plot()
```

# 2 Iterative Policy Evaluation for Q

For deterministic policy $\pi$ and action-independent reward $R$
\begin{array}{lll}
Q^{\pi}(s,a)&=&R(s)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\\
&=&R(s)+\gamma\sum_{s'}P(s,a,s')Q^{\pi}(s',\pi(s'))
\end{array}

For non-deterministic policy $\pi$ and action-independent reward $R$ (this is implemented)
\begin{array}{lll}
Q^{\pi}(s,a)&=&R(s)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\\
&=&R(s)+\gamma\sum_{s'}P(s,a,s')\sum_{a'}\pi(s',a')Q^{\pi}(s',a')
\end{array}

For deterministic policy $\pi$ and action-dependent reward $R$
\begin{array}{lll}
Q^{\pi}(s,a)&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\\
&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')Q^{\pi}(s',\pi(s'))
\end{array}

For non-deterministic policy $\pi$ and action-dependent reward $R$
\begin{array}{lll}
Q^{\pi}(s,a)&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')V^{\pi}(s')\\
&=&R(s,a)+\gamma\sum_{s'}P(s,a,s')\sum_{a'}\pi(s',a')Q^{\pi}(s',a')
\end{array}

```python
from dqn_policy import POLICY
from model_MDP import IPE_Q


" Policy "
policy = POLICY(policy_choice='random')
print(policy.policy)
print()


" Iterative_Policy_Evaluation for Q "
ipe = IPE_Q(policy.policy, num_iteration=500)
ipe.run_ipe()
print(ipe.Q)
print()


" Plot of Q_history "
ipe.plot()
```
